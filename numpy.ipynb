```python
import numpy as np


scores = np.random.randint(0, 100, size=50)

print("Student Scores:\n")
print(scores)



mean_score = np.mean(scores)
median_score = np.median(scores)
highest_score = np.max(scores)
lowest_score = np.min(scores)
std_deviation = np.std(scores)

print("\n===== Statistics =====")
print("Mean Score:", mean_score)
print("Median Score:", median_score)
print("Highest Score:", highest_score)
print("Lowest Score:", lowest_score)
print("Standard Deviation:", std_deviation)



p25 = np.percentile(scores, 25)
p50 = np.percentile(scores, 50)
p75 = np.percentile(scores, 75)

print("\n===== Percentiles =====")
print("25th Percentile:", p25)
print("50th Percentile:", p50)
print("75th Percentile:", p75)




failed_students = scores[scores < 40]


distinction_students = scores[scores > 85]

print("\n===== Failed Students (<40) =====")
print(failed_students)

print("\n===== Distinction Students (>85) =====")
print(distinction_students)



normalized_scores = (scores - np.min(scores)) / (np.max(scores) - np.min(scores))

print("\n===== Normalized Scores =====")
print(normalized_scores)



reshaped_scores = scores.reshape(5, 10)

print("\n===== Reshaped Scores (5x10) =====")
print(reshaped_scores)

subject_averages = np.mean(reshaped_scores, axis=1)

print("\n===== Subject-wise Average Scores =====")
print(subject_averages)



grace_scores = scores + 5

print("\n===== Scores After Adding Grace Marks =====")
print(grace_scores)



theory_marks = np.random.randint(0, 100, size=50)
practical_marks = np.random.randint(0, 100, size=50)

final_scores = (0.4 * theory_marks) + (0.6 * practical_marks)

print("\n===== Theory Marks =====")
print(theory_marks)

print("\n===== Practical Marks =====")
print(practical_marks)

print("\n===== Weighted Final Scores =====")
print(final_scores)


print("\n===== Final Weighted Score Statistics =====")
print("Average Final Score:", np.mean(final_scores))
print("Highest Final Score:", np.max(final_scores))
print("Lowest Final Score:", np.min(final_scores))
```
