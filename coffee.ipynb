```python
from abc import ABC, abstractmethod

class Coffee:
    def __init__(self, name, price):
        self.__name = name
        self.__price = price


    def get_name(self):
        return self.__name

    def get_price(self):
        return self.__price

    @abstractmethod
    def recipe(self):
        pass

    @abstractmethod
    def print_art(self):
        pass


class Latte(Coffee):

    def __init__(self):
        super().__init__("Latte", 180)

    def recipe(self):
        print("\nRecipe for Latte:")
        print("1. Espresso Shot")
        print("2. Steamed Milk")
        print("3. Milk Foam")

    def print_art(self):
        print("""
        ☕ LATTE ☕

        ###########
       ##         ##
      ##  LATTE   ##
      ##   ☕☕☕   ##
       ##         ##
        ###########
        """)


class Matcha(Coffee):

    def __init__(self):
        super().__init__("Matcha", 220)

    def recipe(self):
        print("\nRecipe for Matcha:")
        print("1. Matcha Powder")
        print("2. Hot Milk")
        print("3. Sugar or Honey")

    def print_art(self):
        print("""
        🍵 MATCHA 🍵

        *************
       *             *
      *   MATCHA     *
      *    🍵🍵🍵     *
       *             *
        *************
        """)


class Americano(Coffee):

    def __init__(self):
        super().__init__("Americano", 150)

    def recipe(self):
        print("\nRecipe for Americano:")
        print("1. Espresso Shot")
        print("2. Hot Water")

    def print_art(self):
        print("""
        ☕ AMERICANO ☕

        @@@@@@@@@@@@@
       @@           @@
      @@ AMERICANO @@
      @@    ☕☕☕    @@
       @@           @@
        @@@@@@@@@@@@@
        """)



class CoffeeMachine:

    def serve_coffee(self, coffee):
        print("\n==============================")
        print("Preparing:", coffee.get_name())
        print("Price: ₹", coffee.get_price())
        coffee.print_art()
        coffee.recipe()
        print("==============================")


machine = CoffeeMachine()

while True:

    print("\n===== COFFEE MACHINE =====")
    print("1. Latte")
    print("2. Matcha")
    print("3. Americano")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        coffee = Latte()
        machine.serve_coffee(coffee)

    elif choice == "2":
        coffee = Matcha()
        machine.serve_coffee(coffee)

    elif choice == "3":
        coffee = Americano()
        machine.serve_coffee(coffee)

    elif choice == "4":
        print("Thank you for using Coffee Machine ☕")
        break

    else:
        print("Invalid Choice!")
```
